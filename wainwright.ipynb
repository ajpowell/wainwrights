```python
with open('wainwrights.js', 'w') as fd:
    fd.write('//number,name,heightMetres,heightFeet,latitude,longitude,climbed\n')
    fd.write('var wainwrightLocations = [\n')
    for index, row in df.iterrows():
        fd.write('[{},"{}",{},{},{},{},{}],\n'.format(row['number'], row['name'], row['heightMetres'], row['heightFeet'], row['latitude'], row['longitude'], row['climbed']))
    fd.write(']\n')
```
